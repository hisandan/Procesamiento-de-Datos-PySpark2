### Se puede ver que en muchos casos la edad es nula cuando la persona tiene tarjeta de identidad, como no vamos a utilizar la fecha de nacimiento debido a que no nos sirve para hacer algún análisis estadístico, vamos a utilizar el año de nacimiento y el periodo para obtener la edad aproximada del estudiante

```python
# Extract the year from the 'PERIODO' column
df['PERIODO'] = df['PERIODO'].astype(str).str[:4].astype(int)

# Calculate age by subtracting birth year from the year in 'PERIODO' column
df['EDAD'] = df['PERIODO'] - df['AÑO_NACIMIENTO']


# Calculate the average age
average_age = df['EDAD'].mean()

# Print the average age
print("Edad promedio", average_age)

# Filter the DataFrame for rows where 'ESTU_TIPODOCUMENTO' is equal to 'TI'
filtered_df = df[df['ESTU_TIPODOCUMENTO'] == 'TI']

# Calculate the average age from the filtered DataFrame
average_age_ti = filtered_df['EDAD'].mean()

# Print the average age of people with 'ESTU_TIPODOCUMENTO' equal to 'TI'
print("Edad promedio cuando ESTU_TIPODOCUMENTO = TI:", average_age_ti)

```
